# Lecture 01 - The Learning Problem

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import make_classification
```

## A simple learning algorithm - Perceptron Learning Algorithm

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, class_sep=0.5, flip_y=0.2, random_state=1)
df = pd.DataFrame({'x1': X[:,0], 'x2': X[:,1], 'y':y})
sns.scatterplot(data=df, x='x1', y='x2', hue='y')
```

```python
sign = 1
threshold = df['x1'].mean() + df['x2'].mean()
weight = np.full((2, 1), 1, dtype=int)
df['weight'] = np.full((100, 1), 1, dtype=int)

# This is the perceptron formula
df['perceptron_function'] = sign * np.matmul(X, weight) - threshold


df['predicted_y'] = np.where(df['perceptron_function'] >= 0, 1, 0) 
```

```python
sns.scatterplot(data=df, x='x1', y='x2', hue='predicted_y')
```

After the classification, we will compare the real and predicted output. If the prediction is correct we are done, in other hand, if the prediction is wrong we will calculate a adjusment and sum it to the perceptron function.
